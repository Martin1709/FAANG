```python
import pandas as pd
of1=pd.read_csv(r'C:\Users\ASUS\Desktop\Personal Recent\\FAANG\FAANG.csv')
of1.set_index('Index Names',inplace=True)
of2=of1.iloc[0:1,[0,3,6,9,12]]
#Alternate way of selecting of2=of1.iloc[:,["Facebook FY -2020","Amazon FY -2020","Apple FY -2020","Netflix FY -2020","Google FY -2020"]]
of3=of2.sum(axis=1)
print(of3)
of2.loc['Market_Cap%'] = of2.loc['Market Cap']/4943000000000
print(of2)
```

```python
import matplotlib.pyplot as plt
# Pie chart
labels = ['Facebook ($665B)', 'Amazon ($1.4T)', 'Apple ($1.58T)', 'Netflix ($210B)','Google ($1.04T)']
sizes = of2.loc["Market_Cap%"]
# only "explode" the 2nd slice (i.e. 'Hogs')
explode = (0, 0, 0.1, 0,0)
#add colors
colors = ['#3b5998','#FF9900','#A3AAAE','#E50914','#3cba54']
fig1, ax1 = plt.subplots(figsize=(7,4))
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90,wedgeprops   = { 'linewidth' : 1, 'edgecolor' : "black" },frame=False)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()
plt.show()
fig1.savefig('Pie_Chart')
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 3
Amazon = [232,280,296]

fig, ax = plt.subplots(figsize=(4.5,3))
colors1 =['#0072CE','#FF9900']
ind = np.arange(N)    # the x locations for the groups
width = 0.35         # the width of the bars
p1 = ax.bar(ind, Amazon, width, color=colors1[1],alpha=0.8)



Walmart = [504,514,523]
p2 = ax.bar(ind + width, Walmart, width,color=colors1[0],alpha=0.8)

ax.set_title('Revenues - Amazon vs Walmart')
ax.set_xticks(ind + width / 2)
ax.set_ylabel('$ Billions')
ax.set_xticklabels(('FY 2017-18', 'FY 2018-19', 'FY 2019-20'))

ax.legend(('Amazon', 'Walmart'))
ax.autoscale_view()

plt.show()
```
